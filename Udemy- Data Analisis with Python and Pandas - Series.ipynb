## Lesson 42 The .apply() Method

```python
google_stock_price = pd.read_csv('google_stock_price.csv', squeeze=True)
```

```python
def classify_perfomance(number):
    if number < 300:
        return 'OK'
    elif number >= 300 and number < 650:
        return 'Satisfactory'
    else:
        return 'Incredible'
```

```python
google_stock_price.apply(classify_perfomance)
```

```python
google_stock_price.head()
```

```python
google_stock_price.apply(lambda stock_price : stock_price + 1)
```

## Lesson 43 The .map() Method

```python
pokemon_names = pd.read_csv('pokemon.csv', usecols=['Pokemon'], squeeze=True)
pokemon_names.head(3)
```

```python
pokemon_types = pd.read_csv('pokemon.csv', index_col='Pokemon', squeeze=True)
pokemon_types.head(3)
```

```python
pokemon_names.map(pokemon_types)
```

```python
pokemon_names = pd.read_csv('pokemon.csv', usecols=['Pokemon'], squeeze=True)
pokemon_types = pd.read_csv('pokemon.csv', index_col='Pokemon', squeeze=True).to_dict()
```

```python
pokemon_names.map(pokemon_types)
```
